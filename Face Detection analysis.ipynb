```python
import os
import cv2

# function to create dictionary of paths for each celebrity
def get_folder_contents(root_folder):
    folder_dict = {}

    # Iterate through immediate subdirectories of the root folder
    for dir_name in os.listdir(root_folder):
        dir_path = os.path.join(root_folder, dir_name)
        if os.path.isdir(dir_path):
            folder_dict[dir_name] = [os.path.join(dir_path, f) for f in os.listdir(dir_path)]

    return folder_dict

# Function to encode cropped image
def encode_image(path, target=(64, 64)):
   
    image = cv2.imread(path)
    if image is not None:
        # raise ValueError("Unable to load image at path: {}".format(path))
        
        # Resize the image to the target size
        image = cv2.resize(image, target)
        
        # Convert the image to RGB (OpenCV loads images as BGR by default)
        # image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        
        
        # Normalize pixel values to be between 0 and 1
        image_normalized = image_rgb / 255.0
        # image_normalized = image_gray / 255.0
        # Add a batch dimension
        image_expanded = np.expand_dims(image_normalized, axis=0)
        # image_expanded = np.expand_dims(image_rgb, axis=0)
        return image_expanded
        # return image_gray
        # return image_rgb
        
```

```python
import pandas as pd
import numpy as np

# creating a dictionary of paths for cropped folder
celebrity_file_names_dict = get_folder_contents("./cropped - Copy")


#creating key mapping dictionary for celebrity name
y_count = 1
y_dict = {}
for i in celebrity_file_names_dict.keys():
    y_dict[i] = y_count
    y_count += 1
y_dict

# Function for getting encoded value for a sportsman person
def celebrity_code(name):
    return y_dict.get(name)


# creating a dataframe for all images for eacch celebrity
encoded_df = pd.DataFrame()
celeb_df = pd.DataFrame()
celeb = []

for celebrity, celebrity_image_folder in celebrity_file_names_dict.items():
    # print(celebrity_image_folder)
    for img in celebrity_image_folder:
        en_img = encode_image(img)
        if en_img is not None:
            encoded_df = pd.concat([encoded_df, pd.DataFrame(en_img.reshape(1, -1))], ignore_index=True)
            celeb.append(celebrity_code(celebrity))
    celeb_df = pd.DataFrame(celeb, columns=["celeb"])



# # x = encode_image("./cropped/john cena/john cena1.png")
# # y = encode_image("./cropped/john cena/john cena2.png")

# # df_x = pd.DataFrame(x.reshape(1, -1))
# # df_y = pd.DataFrame(y.reshape(1, -1))

# # df = pd.concat([df_x, df_y], ignore_index=True)


# df = pd.DataFrame()
df = pd.concat([encoded_df, celeb_df], axis = "columns")
# # df.to_csv("check.csv")

# encoded_df
# celeb_df
df
```

```python
#testing
import cv2

# making classifier to decect front face and 2 eyes
face_cascade = cv2.CascadeClassifier('./haarcascade/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('./haarcascade/haarcascade_eye.xml')


# function to crop image as per classifiers
def crop_image(path):
    # Read the image
    img = cv2.imread(path)
    
    # Check if image is empty or invalid
    if img is None:
        print("Error: Unable to read the image file.")
        return None
    
    # Convert the image from color to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Check if the grayscale image is empty or invalid
    if gray is None:
        print("Error: Unable to convert the image to grayscale.")
        return None
    face = face_cascade.detectMultiScale(gray, 1.3, 5)    # detects faces in the grayscale image
    
    for (x,y,w,h) in face:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >= 2:
            # return roi_color
            roi_color = cv2.resize(roi_color, (64,64))
            image_rgb = cv2.cvtColor(roi_color, cv2.COLOR_BGR2RGB)
        
            image_normalized = image_rgb / 255.0
            image_expanded = np.expand_dims(image_normalized, axis=0)
            return image_expanded.reshape(1, -1)
            
# input_image_path = "A:\Education\Data Science\Machine Learning\Projects\Image Classification\model\MESSI.jpg"
input_image_path = "./MESSI.jpg"
print(crop_image(input_image_path))
predict = model.predict(crop_image(input_image_path))[0]
# print(predict)
# print(y_dict[predict])
def get_key_from_value(dictionary, value):
    return next((key for key, val in dictionary.items() if val == value), None)
print(get_key_from_value(y_dict, predict))
```
